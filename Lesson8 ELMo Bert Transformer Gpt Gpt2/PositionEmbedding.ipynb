### 根据 Transformer 中对位置编码 Position Embedding 的公式描述，似乎 位置编码只需要计算一次就行了。

比如：句子最长是200个词， 词向量维度是700, 那么可以求出 Position Embedding 的结果张量：

```python
import numpy

sentence_max_length = 10
d_word_vector = 20
posList = numpy.arange(0, sentence_max_length)
iList = numpy.arange(0, d_word_vector)

pos_emb = numpy.zeros((sentence_max_length, d_word_vector))
x = []
y = []
z = []
for pos in posList:
    for i in iList:
        x.append(pos)
        y.append(i)
        if i % 2 == 0:
            pos_emb[pos, i] = numpy.sin(pos/numpy.power(10000, 2 * i / d_word_vector)) 
        else:
            pos_emb[pos, i] = numpy.cos(pos/numpy.power(10000, 2 * i / d_word_vector)) 
        z.append(pos_emb[pos, i])

print((pos_emb * 1000).astype(int))
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

plt.rcParams['figure.figsize'] = (15.0, 15.0) # 设置figure_size尺寸

ax1 = plt.axes(projection='3d')
ax1.scatter3D(x, y, z ) #绘制散点图
ax1.plot3D(x, y, z) #绘制散点图
```
