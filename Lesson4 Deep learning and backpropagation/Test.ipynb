```python
x = torch.linspace(-6.0, 6, 270)
# x = torch.unsqueeze(x, dim=1)
x = Variable(x)
# print(x.size())
y = torch.pow(x,2) + 5*torch.rand(x.size())
y = Variable(y)
# print(x, y)

pyplot.scatter(x.data.numpy(), y.data.numpy())
pyplot.show()
```

```python
n_data = torch.ones(100, 2)
x0 = torch.normal(2*n_data, 1)
y0 = torch.zeros(100)
x1 = torch.normal(-2*n_data, 1)
y1 = torch.ones(100)
x = torch.cat((x0,x1),0)
y = torch.cat((y0, y1)).type(torch.long)
x, y
```

```python
class Net(torch.nn.Module):
    def __init__(self, featureCount, hittenLayerCount, outputCont):
        super(Net, self).__init__()
        self.hiddenLayer = torch.nn.Linear(featureCount, hittenLayerCount)
        self.predictLayer = torch.nn.Linear(hittenLayerCount, outputCont)
        
    def forward(self, x, activationFunction1, activationFunction2):
        return activationFunction2(self.predictLayer(activationFunction1(self.hiddenLayer(x))))
```

```python
net = Net(2,10,2)

optimizer = torch.optim.SGD(net.parameters(), lr = 0.1)
lossFunc = torch.nn.CrossEntropyLoss()

for i in range(100):
    predictData = net.forward(x, torch.nn.functional.tanh, torch.nn.functional.softmax)
    
    # optimizer.zero_grad()
    lossFunc(predictData, y,).backward()
    optimizer.step()
    
print("predictData",predictData)
```
