```python
# 爬起地铁数据：https://map.bjsubway.com/ 

import requests
 
response = requests.get('https://map.bjsubway.com/subwaymap/beijing.xml?v=2')

print(response.text)
```

```python
from bs4 import BeautifulSoup 
be = BeautifulSoup(response.text,"lxml")
```

```python
print(type(be))
lines = [p for p in be.find_all('l')]
line_map = {}
port_map = {}
for l in lines:
    ports = l.find_all('p')
    port_line = line_map[l['lid']] = []
    is_loop = l['loop'] == 'true'
    print(l['lid'], "is_loop:",is_loop)
    for port in ports:
        if port['acc'] == '': continue
#         if not port.has_key('ax'): continue
        port_map[port['lb']] = (float(port['x']), float(port['y']), 
                                float(port['rx']), float(port['ry']), 
#                                 float(port['ax']), float(port['ay'])
                               )
        port_line.append(port['lb'])
    if is_loop:
        port_line.append(port_line[0])
    for port in port_line:
        print('-'*10, port)
```

```python
import networkx
# import matplotlib.pyplot as plt


all_ports1 = {}
for lineName, ports in line_map.items():
    for port in ports:
        all_ports1[port] = (port_map[port][0], -port_map[port][1])
    
```

```python
import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['SimHei'] 
matplotlib.rcParams['font.family']='sans-serif'
matplotlib.rcParams['figure.figsize'] = (32.0, 18.0)
```

```python
city_graph1 = networkx.Graph()
city_graph1.add_nodes_from(all_ports1.keys())
networkx.draw(city_graph1, all_ports1, with_labels=True, node_size=10)
```

```python
from collections import defaultdict
```

```python
# 建立连接，这里假设字典是按顺序遍历的，若是不按照存入顺序遍历的话，这里将建立错误的连接
connections = defaultdict(list)
for lineName, ports in line_map.items():
    last = None
    for port in ports:
        if last != None:
            connections[port].append(last)
            connections[last].append(port)
        last = port
connections
```

#### 3. Build the search agent

Build the search agent based ont he graph you built.

for example, if you use Beijing subway graoh, and you run:

\>>> search("奥体中心“，”天安门“）

You should get the result as follows: 奥体中心 -> A ->B ->C ... -> 天安门

```python
def show_path(path):
    print( ' -> '.join(path))

show_path(['奥体中心','a','b','c','天安门'])
```

```python
def search_DFS(graph, startPort, endPort, line = []):
    if startPort in graph and endPort in graph:
        line.append(startPort)
        for port in graph[startPort]:
            if port in line:
                continue
            if port == endPort:
                line.append(endPort)
                return line, True
#             print(port)
            new_line, isFound = search_DFS(graph, port, endPort, list.copy(line))
            if isFound:
                return new_line, True
        return None, False
        
my_line, is_found = search_DFS(connections, '九龙山', '群芳')
show_path(my_line)
```

```python
def search_BFS(graph, startPort, endPort, search_strategy = lambda x:x):
    if startPort in graph and endPort in graph:
        steps = set()
        paths = [[startPort]]
        while paths:
            paths = search_strategy(paths) # 策略放在这里，若这里排序，下面取出最优的路径
            # 减少 paths，因为后面是从后面增加，所以这里要取出第0个，
            # 如果这里不填0，而是空着，那么将从最后面取出，那么这个方法将变成深度优先算法
            path = paths.pop(0) 
#             print(path)
            last_port = path[-1]
            
            if last_port == endPort:
                return path
            
#             print(last_port in graph)
            if last_port in graph:
                for port in graph[last_port]: # 如果还有子节点，则扩张 paths
                    if port in steps:
                        continue
                    steps.add(port)
                    new_path = list.copy(path)
                    new_path.append(port)
#                     print(new_path)
                    paths.append(new_path) 
            
            
my_line = search_BFS(connections, '九龙山', '群芳')
show_path(my_line)
```
